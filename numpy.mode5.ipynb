```python
import numpy as np
import pandas as pd
```

```python
temperatures_w1 = np.array([22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9])
print("Week 1 Temperatures:", temperatures_w1)
```

```python
print("Shape:", temperatures_w1.shape)
print("Data Type:", temperatures_w1.dtype)
print("Number of Elements:", temperatures_w1.size)
```

```python
temperatures_f = (temperatures_w1 * 9/5) + 32
print("Temperatures in Fahrenheit:", temperatures_f)
```

```python
print("Maximum Temperature:", temperatures_w1.max())
print("Minimum Temperature:", temperatures_w1.min())
print("Mean Temperature:", temperatures_w1.mean())
```

```python
# First three days
print("First 3 days:", temperatures_w1[:3])

# Weekend (last two days)
print("Weekend temperatures:", temperatures_w1[-2:])

# Middle three days
print("Middle 3 days:", temperatures_w1[2:5])
```

```python
temperatures = np.array([
    [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9],
    [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]
])

print("2D Temperature Array:\n", temperatures)
```

```python
print("Shape:", temperatures.shape)
print("Data Type:", temperatures.dtype)
print("Total Elements:", temperatures.size)

# Temperatures for each week
print("Week 1:", temperatures[0])
print("Week 2:", temperatures[1])

# Weekend temperatures for both weeks
print("Weekend temperatures:\n", temperatures[:, -2:])
```

```python
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

print("Marks Series:\n", marks)
```

```python
#Indexing and Slicing
#1st Rank (Integer Index Position)

print("1st Rank Mark (by position):", marks.iloc[0])
```

```python
#Top 3 Ranks using loc

top3 = marks.loc[['Rank1', 'Rank2', 'Rank3']]
print("Top 3 Ranks Marks:")
print(top3)
```

```python
#3rd Rank using iloc
print("3rd Rank Mark (using iloc):", marks.iloc[2])
```

```python
#Boolean Mask (Marks > 90)
above_90 = marks[marks > 90]
print("Marks greater than 90:")
print(above_90)
```

```python
#Manipulating Series Modify 1st Rank to 100
marks.loc['Rank1'] = 100
print("After Updating Rank1:")
print(marks)
```

```python
#Remove Last Rank (Rank5)

marks = marks.drop('Rank5')
print("After Removing Rank5:")
print(marks)
```

```python
#Compute CGPA (Marks / 10)

cgpa = marks / 10
print("CGPA:")
print(cgpa)
```

```python
#Creating Pandas DataFrame

transactions = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': [
        'Electronics', 'Clothing', 'Electronics', 'Furniture',
        'Clothing', 'Electronics', 'Furniture', 'Clothing',
        'Furniture', 'Electronics'
    ],
    'Region': [
        'North', 'South', 'North', 'East', 'West',
        'North', 'East', 'West', 'South', 'North'
    ],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})

print(transactions)
```

```python
#Data Exploration
# Head, Tail, Shape, Columns, Data Types

print("Head:")
print(transactions.head())

print("\nTail:")
print(transactions.tail())

print("\nShape:", transactions.shape)

print("\nColumn Names:")
print(transactions.columns)

print("\nData Types:")
print(transactions.dtypes)

print("\nInfo():")
print(transactions.info())
```

```python
#Display Only 'ProductCategory' and 'Amount'

print(transactions[['ProductCategory', 'Amount']])
```

```python
#Retrieve Last 3 Columns (using iloc)

last_3_columns = transactions.iloc[:, -3:]
print("Last 3 Columns:")
print(last_3_columns)
```

```python
#Filter: Region = 'North' AND Amount > 200

north_high_amount = transactions[
    (transactions['Region'] == 'North') & (transactions['Amount'] > 200)
]

print("North Region with Amount > 200:")
print(north_high_amount)
```

```python
#Value Counts for 'ProductCategory'

product_counts = transactions['ProductCategory'].value_counts()
print("Product Category Counts:")
print(product_counts)
```

```python
#Unique Values in 'Region'

unique_regions = transactions['Region'].unique()
print("Unique Regions:")
print(unique_regions)
```

```python
#Group by 'Region' and Mean Amount

region_mean = transactions.groupby('Region')['Amount'].mean()
print("Mean Amount by Region:")
print(region_mean)
```

```python
#Manipulating the DataFrame
#Modify Amount for TransactionID = 102 to 165

transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165
print("After Updating TransactionID 102:")
print(transactions)
```

```python
#Add New Column 'Discount' (10% of Amount)

transactions['Discount'] = transactions['Amount'] * 0.10
print("After Adding Discount Column:")
print(transactions)
```

```python
#Remove Row with TransactionID = 109

transactions = transactions[transactions['TransactionID'] != 109]
print("After Removing TransactionID 109:")
print(transactions)
```

```python
transactions.drop(columns='Discount', inplace=True)
```
